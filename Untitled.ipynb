```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
COMBINED = pd.read_csv('Hemisphere_Combined_Data_Sets.py.csv')
COMBINED.head()
```

```python
plt.plot(COMBINED.Number, COMBINED.Corona_Confirmed, color='Tan')
plt.plot(COMBINED.Number, COMBINED.Corona_Death, color = 'Blue')
plt.show()
```

```python
# Add labels to X and Y axes :: Add title
plt.title("COVID-19 Contractions v.s. Deaths")
plt.xlabel("Days")
plt.ylabel("Total Global Cases per Day")
plt.xlim(0,55)
plt.ylim(0, 70000, 10000)
plt.legend

Tan_patch = mpatches.Patch(color='Gold', label='COVID-Confirmed ')
Blue_patch = mpatches.Patch(color='Red', label='COVID-Death')
plt.legend(handles=[Tan_patch, Blue_patch])


plt.plot(COMBINED.Number, COMBINED.Corona_Confirmed, color='Gold')
plt.plot(COMBINED.Number, COMBINED.Sars_Confirmed, color = 'Red')
plt.show()
```

```python
plt.plot(COMBINED.Number, COMBINED.Sars_Confirmed, color='Tan')
plt.plot(COMBINED.Number, COMBINED.Sars_Death, color = 'Blue')
plt.show()
```

```python
# Add labels to X and Y axes :: Add title
plt.title("Sars Contractions v.s. Deaths")
plt.xlabel("Days")
plt.ylabel("Total Global Cases per Day")
plt.xlim(0,55)
plt.ylim(0, 6000, 1000)
plt.legend

Tan_patch = mpatches.Patch(color='Silver', label='Sars-Confirmed ')
Blue_patch = mpatches.Patch(color='Red', label='Sars-Death')
plt.legend(handles=[Tan_patch, Blue_patch])


plt.plot(COMBINED.Number, COMBINED.Sars_Confirmed, color='Silver')
plt.plot(COMBINED.Number, COMBINED.Sars_Death, color = 'Red')
plt.show()
```
